```python
import pickle

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
all_ea_seeds = []
for seed in np.arange(2024, 2034):
    with open(f'../data_shap_values/SHAP_values_EA_final_{seed}.pkl', 'rb') as f:
        values_ea = pickle.load(f)
    # find location where pixel values are above the mean

    mean_ea = np.abs(values_ea[1].flatten()).mean()
    std_ea = np.abs(values_ea[1].flatten()).std()
    
    all_ea = []
    for i in range(values_ea[1].shape[0]):
        above_ = np.where(np.abs(values_ea[1][i, :, :, 0].flatten()) > mean_ea + 2*std_ea)[0]
        if above_.shape[0] > 0:
            all_ea.append(above_.shape[0])
    all_ea_seeds.extend(all_ea)
```

```python
len(all_ea_seeds), mean_ea, std_ea
```

```python
fig, ax = plt.subplots(1, 1, figsize=(4, 3))
sns.ecdfplot(data = all_ea_seeds, 
             stat='proportion', ax=ax, color= '#C25759')

ax.set(ylim=(0, 1))
ax.set(xlim=(0, 250))
ax.set_xlabel('Pixels')
# ax.legend(['DAiSEE', 'DevEmo', 'EA'])
# fig.savefig('../figure_panels/EA_SHAP_CDF_all.svg', bbox_inches='tight', dpi=300)
```

```python
all_indian_seeds = []
for seed in np.arange(2024, 2034):
    with open(f'../data_shap_values/SHAP_values_DAiSEE_final_{seed}.pkl', 'rb') as f:
        values_indian = pickle.load(f)
    # find location where pixel values are above the mean

    mean_indian = np.abs(values_indian[1].flatten()).mean()
    std_indian = np.abs(values_indian[1].flatten()).std()

    
    all_indian = []
    for i in range(values_indian[1].shape[0]):
        # above_ = np.where(np.abs(values_indian[1][i, :, :, 0].flatten()) > mean_indian + 2*std_indian)[0]
        above_ = np.where(values_indian[1][i, :, :, 0].flatten() > mean_indian + 2*std_indian)[0]
        if above_.shape[0] > 0:
            all_indian.append(above_.shape[0])
    all_indian_seeds.extend(all_indian)
```
