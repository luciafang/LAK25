```python
# load pickle file
files = ['../data_features_labels/Train_imb.pkl', '../data_features_labels/Test_imb.pkl', '../data_features_labels/Validation_imb.pkl', 
         ]
# files = ['Eastern_Asian_Dataset.pkl']
x_train_list = []
y_train_list = [] 
for file in files:
    with open(file, 'rb') as f:
        features, labels = pickle.load(f)
    
    for pic, label in zip(features, labels):
        # plt.imshow(img_gray, cmap='binary_r')
        x_train_list.append(pic)
        y_train_list.append(label)
```
